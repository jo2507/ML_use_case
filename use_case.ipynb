```python
f={'Experience':[1,2,3.6,4,5,6,7,8],
   'Salary':[4,5,5.5,8,9,11,12.5,14]
   }
df=pd.DataFrame(f)
df
```

```python
x=df['Experience']
y=df['Salary']
```

```python
# Sample data
X = [1, 2, 3.6, 4, 5, 6, 7, 8]
Y = [4, 5, 5.5, 8, 9, 11, 12.5, 14]

# initialize slope (m) and intercept (c)
m = 0
c = 0

# learning rate
lr = 0.01

# number of iterations
epochs = 1000

n = len(X)

# Gradient Descent
for i in range(epochs):
    y_pred = []
    for j in range(n):
        y = m * X[j] + c
        y_pred.append(y)

    # calculate gradients
    dm = 0
    dc = 0
    for j in range(n):
        dm += -2 * X[j] * (Y[j] - y_pred[j])
        dc += -2 * (Y[j] - y_pred[j])

    # update m and c
    m = m - (lr * dm / n)
    c = c - (lr * dc / n)

# after training
print("Optimized slope (m):", round(m, 3))
print("Optimized intercept (c):", round(c, 3))

# Predictions
print("\nPredictions:")
for i in range(n):
    pred = m * X[i] + c
    print(f"Exp: {X[i]} -> Actual: {Y[i]}  Predicted: {round(pred, 2)}")

# ---- Evaluation Metrics ----
mse = 0
mae = 0
for i in range(n):
    mse += (Y[i] - (m * X[i] + c))**2
    mae += abs(Y[i] - (m * X[i] + c))

mse = mse / n
rmse = mse ** 0.5
mae = mae / n

# R2 Score
mean_y = sum(Y)/n
ss_tot = sum([(y - mean_y)**2 for y in Y])
ss_res = sum([(Y[i] - (m*X[i] + c))**2 for i in range(n)])
r2 = 1 - (ss_res/ss_tot)

print("\nMSE:", round(mse, 3))
print("RMSE:", round(rmse, 3))
print("MAE:", round(mae, 3))
print("R2 Score:", round(r2, 3))

# ---- Visual Part ----
import matplotlib.pyplot as plt
preds=[]
for i in range(len(x)):
  preds.append(m*x[i] + c)

plt.scatter(X, Y, color='blue', label='Actual line')   # actual points
plt.plot(X,preds, color='orange', label='Predicted Line') # regression line

plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Actual vs Predicted Salary')
plt.legend()
plt.grid(True)
plt.show()
```
